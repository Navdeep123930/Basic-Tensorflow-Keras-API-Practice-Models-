Multivariate Linear Regression

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.optimizers import RMSprop
import numpy as np
x=np.array([[1,2],[2,4],[3,6],[4,8],[5,10],[6,12],[7,14],[8,16],[9,18]])
y=np.array([[4],[8],[12],[16],[20],[24],[28],[32],[36]])
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=(2,))])
model.compile(optimizer=RMSprop(0.05),loss='mse')
model.fit(x,y,epochs=500)
```

```
print(model.predict([[10,20]]))
```
